```python
df_all = pd.DataFrame()
```

```python
import os
```

```python
for m in ms:
    fname = '../sf/V827Tau/m{}/output/mix_emcee/run01/models_ff-05_50_95.csv'.format(m)
    if os.path.isfile(fname): 
        print(m)
        df = pd.read_csv(fname)
        df['m_order'] = m
        df_all = df_all.append(df, ignore_index=True)
```
