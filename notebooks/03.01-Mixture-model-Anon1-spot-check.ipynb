# Spotcheck mixture model of Anon1 

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import numpy as np
```

```python
df_comp = pd.read_csv('../sf/Anon1/m112/output/mix_emcee/run01/spec_config_composite.csv')
df_cool = pd.read_csv('../sf/Anon1/m112/output/mix_emcee/run01/spec_config_cool.csv')
df_hott = pd.read_csv('../sf/Anon1/m112/output/mix_emcee/run01/spec_config_hot.csv')
```

```python
df = pd.read_csv('../sf/Anon1/m112/output/mix_emcee/run01/spec_config.csv')
```

```python
CC = np.load('../sf/Anon1/m112/output/mix_emcee/run01/CC.npy')
```

```python
draw = np.random.multivariate_normal(df.model_composite,CC)
```

```python
plt.step(df.wl, df.data, 'k-')

cool = 4.0*df_cool.model_composite
hot = 0.6*df_hott.model_composite
net = cool+hot

#plt.plot(df.wl, net, label='composite')
plt.plot(df.wl, cool, label='cool')
plt.plot(df.wl, hot, label='hot')
plt.plot(df.wl, draw, label='draw')
plt.plot(df.wl, df.model_composite, label='composite')

plt.legend(loc='best')
plt.xlabel('$\lambda \;(\AA)$')
#plt.ylim(0,1.2)
```
